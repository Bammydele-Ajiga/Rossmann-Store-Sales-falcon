<h2> DATA PREPROCESSING</h2>

```python
train_df
```

```python
#DROP UNNESSARY COLUMNS
# train_df.drop(["Customers"], axis =1, inplace = True) #Number of customers not in test set

train_df.drop(["Dates","Date", "Year"], axis = 1, inplace = True) # Not useful in training
```

```python
#Day of week shouldnt have diffent weight, therefore one hot encoding is applied to create dummmy variables.

#One hot encoding of 
day_dummies = pd.get_dummies(train_df['DayOfWeek'],prefix = "Day")

#Join day_dummies to train_df

train_df = train_df.join(day_dummies)
```

```python

train_df.drop(["DayOfWeek"], inplace = True, axis = 1)
```

```python
train_df
```

```python
test_day_dummies = pd.get_dummies(test_df['DayOfWeek'],prefix = "Day")

test_df = test_df.join(test_day_dummies)

test_df.drop(["DayOfWeek"], inplace = True, axis = 1)

test_df.drop(["Date"], inplace = True, axis = 1)

test_df
```

```python
#M
test_df.set_index('Id',inplace = True)

test_df
```

```python
#MERGE TRAIN AND TEST DATASET WITH

merge_train_set =  pd.merge(left=train_df,
                               right=store_df,
                               on='Store',
                               how='left')
```

```python
merge_train_set
```

```python
merge_test_set  =  pd.merge(left=test_df,
                               right=store_df,
                               on='Store',
                               how='left')
```

```python
merge_test_set
```

```python
#One Hot Encoding of Store type



storeType_dummies = pd.get_dummies(merge_train_set['StoreType'],prefix = "StoreType")

merge_train_set = merge_train_set.join(storeType_dummies)

merge_train_set
```

```python
merge_test_set  =  pd.merge(left=test_df,
                               right=store_df,
                               on='Store',
                               how='left')
storeType_dummies = pd.get_dummies(merge_test_set['StoreType'],prefix = "StoreType")

merge_test_set = merge_test_set.join(storeType_dummies)

```

```python
merge_test_set
```

```python
merge_train_set.drop(["StoreType"], axis =1, inplace = True)

merge_test_set.drop(["StoreType"], axis =1, inplace = True)
```

```python
#label Enconding of Assortment, Its save to label encode because Assortment - describes an assortment level: a = basic, b = extra, c = extended

merge_train_set["Assortment"] = merge_train_set["Assortment"].map({"a": 1, "b": 2, "c": 3})

merge_test_set["Assortment"] = merge_test_set["Assortment"].map({"a": 1, "b": 2, "c": 3})


merge_test_set
```

```python
#Drop customer columns as this is not present in test data set

merge_train_set.drop(["Customers"], inplace =True, axis = 1)
```

We have our dataset cleaned, so we can start training and making decision.


```python
#EXproting train and test dataset

merge_train_set.to_csv("merge_train_set.csv")

```
