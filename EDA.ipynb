<h2> Average Sales per month</h2>

```python
#Create new Column "Dates" with year and month from Date column
train_df['Dates'] = train_df['Date'].apply(lambda x: (str(x)[:7]))

train_df
```

```python
monthly_average_sales  = train_df.groupby('Dates')["Sales"].mean()

# # plot average sales over time(year-month)
ax1 = plt.figure(figsize=(15,6))
ax1 = monthly_average_sales.plot(legend=True,marker='o',title="Average Sales")

ax1.set_xticks(range(len(monthly_average_sales)))
ax1.set_xticklabels(monthly_average_sales.index.tolist(), rotation=90)

# plt.xlabels(monthly_average_sales.index.tolist(), rotation=45)

```
